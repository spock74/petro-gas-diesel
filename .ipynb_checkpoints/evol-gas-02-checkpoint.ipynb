# Análise da evolução do preço da Gasolina, Etanol e Óleo Diesel

## Julho 2001 a Abril 2018

----

```r
suppressPackageStartupMessages( library(dplyr) )
suppressPackageStartupMessages( library(ggplot2) )
suppressPackageStartupMessages( library(plotly) )
suppressPackageStartupMessages( library(readr) )
```

```r
#dd <- readRDS('IN_W_LEAP_YEAR_2001_2018.rds')
dd <- read_csv("IN_W_LEAP_YEAR_2001-2018.csv", col_types = cols(MES = col_character()))
dd$MES <- as.Date(dd$MES)
dd$PRODUTO <- as.factor(dd$PRODUTO)
```

```r
glimpse(dd)
```

```r
stopifnot( isTRUE( nrow(dplyr::filter(dd, PRODUTO == "ETANOL_HIDRATADO")) == 202 ))
stopifnot( isTRUE( nrow(dplyr::filter(dd, PRODUTO == "GASOLINA")) == 202 ) )
stopifnot( isTRUE( nrow(dplyr::filter(dd, PRODUTO == "GLP")) == 202 ) )
stopifnot( isTRUE( nrow(dplyr::filter(dd, PRODUTO == "GNV")) == 202 ) )
#
```

```r
gas <- dplyr::filter(dd, PRODUTO == "GASOLINA")
etanol <- dplyr::filter(dd, PRODUTO == "ETANOL_HIDRATADO")
diesel <- dplyr::filter(dd, PRODUTO == "OLEO_DIESEL")
gnv <- dplyr::filter(dd, PRODUTO == "GNV")
```

```r
dd <- rbind(gas, etanol, diesel, gnv)
```

```r
ggplot(dd, aes(x = MES, y = PRECO_MEDIO_REVENDA, colour = PRODUTO)) + geom_line() 
```

```r
p1 <- ggplot(dd, aes(x = MES, y = PRECO_MEDIO_REVENDA, colour = PRODUTO)) + geom_line() 
```
